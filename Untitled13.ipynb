```python
#1)generating matric marksheet
student_name=str(input("enter your name: "))
roll_no=str(input("enter your roll_no: "))
board=str(input("enter your affiliated board: "))

#obtain marks and total marks
obtain_marks=int(input("enter obtain marks: "))
total_marks=int(input("enter total marks "))
#calculating percentage
percentage= (obtain_marks/total_marks)*100
#grading policy
if percentage >= 90:
    grade = 'A1'
elif percentage >= 80:
    grade = 'A+'
elif percentage >= 70:
    grade = 'A'
elif percentage >= 60:
    grade = 'B'
elif percentage >= 50:
    grade =  'C'
elif percentage >= 40:
    grade =  'D'
else:
    grade = 'f'
#calculating matric marks and generating marksheet
print("---SSC MARKSHEET---")
print("according to your information here is your ssc marksheet")
print("student name:",student_name)
print("roll no:",roll_no)
print("affliated board:",board)
print("obtained marks:",obtain_marks)
print("total marks:",total_marks)
print("percentage:",percentage)
print("Grade:",grade)
print("now time to generate inter marksheet press enter to continue")
# 2)generating inter marksheet HSSC
student_name=str(input("enter your name: "))
roll_no=str(input("enter your roll_no: "))
board=str(input("enter your affiliated board: "))
group=str(input("enter your group: "))

#obtain marks and total marks
obtain_marks=int(input("enter obtain marks: "))
total_marks=int(input("enter total marks:"))
#calculating percentage
percentage= (obtain_marks/total_marks)*100
#grading policy
if percentage >= 90:
    grade = 'A1'
elif percentage >= 80:
    grade = 'A+'
elif percentage >= 70:
    grade = 'A'
elif percentage >= 60:
    grade = 'B'
elif percentage >= 50:
    grade =  'C'
elif percentage >= 40:
    grade =  'D'
else:
    grade = 'f'
#calculating matric marks and generating marksheet
print("----HSSC MARKSHEET---")
print("here is your hssc marsheet")
print("student name:",student_name)
print("roll no:",roll_no)
print("affliated board:",board)
print("Group:",group)
print("obtained marks:",obtain_marks)
print("total marks:",total_marks)
print("percentage:",percentage)
print("Grade:",grade)
```
